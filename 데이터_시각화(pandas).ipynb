여러개의 그래프를 생성하는 방법 (subplot)

subplot(row(행),column(열),index)


```python
data = np.arange(100,201)
plt.subplot(2,1,1) # 2 1 1의 뜻은 1개가 아니라 2개의 행 1개의 열 중에 1번 그래프 
plt.plot(data)

data2 = np.arange(201,301)
plt.subplot(2,1,2)  # 2 1 1의 뜻은 1개가 아니라 2개의 행 1개의 열 중에 2번 그래프 
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100,201)
plt.subplot(211) #컴마 안 써도 된다. 
plt.plot(data)

data2 = np.arange(201,301)
plt.subplot(212) 
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100,201)
plt.subplot(1,3,1)  
plt.plot(data)

data2 = np.arange(201,301)
plt.subplot(1,3,2) 
plt.plot(data2)

data3 = np.arange(301,400)
plt.subplot(1,3,3) 
plt.plot(data3)


plt.show()
```

여러개의 plot을 그리는 방법 (sudplot)+s

plt.sudplots(행의개수,열의개수)

```python
data = np.arange(1,51)

fig, axes = plt.subplots(2,3)

axes[0,0].plot(data)
axes[0,1].plot(data * data)
axes[0,2].plot(data ** 3)
axes[1,0].plot(data % 10)
axes[1,1].plot(-data)
axes[1,2].plot(data // 20)

plt.tight_layout()
plt.show()

```
